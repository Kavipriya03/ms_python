```python
!wget = https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv
df_students=pd.read_csv('grades.csv',delimiter=',',header='infer')
df_students.head()
```

```python
df_students.isnull()
```

```python
df_students.isnull().sum()
```

```python
#coloum is axis 1 and row is axis 2
df_students[df_students.isnull().any(axis=1)]
```

```python
df_students.StudyHours = df_students.StudyHours.fillna(df_students.StudyHours.mean())
df_students
```

```python
#dropping empty values we use how='any' because any values that has NaN will be dropped
df_students=df_students.dropna(axis=0,how='any')
df_students
```

```python
#to get column name as index
mean_study=df_students['StudyHours'].mean()
#to get column name as property
mean_grade=df_students.Grade.mean()
print('Average mean study hours {:2f}\nAverage mean grade {:2f}'.format(mean_study,mean_grade))
```

```python
#students who studied for more or equal to mean hours
df_students[df_students.StudyHours>mean_study]
```

```python
# Get students who studied for the mean or more hours
df_students[df_students.StudyHours > mean_study]
```

```python
passes = pd.Series(df_students['Grade']>=60)
df_students=pd.concat([df_students, passes.rename("Pass")],axis=1)
df_students
```

```python
print(df_students.groupby(df_students.Pass).Name.count())
```

```python
print(df_students.groupby(df_students.Pass)[['StudyHours', 'Grade']].mean())
```

```python
#creating subplots inside the existing plot
fig, ax = plt.subplots(1,2 ,figsize=(10,4))
ax[0].bar(x=df_students.Name,height=df_students.Grade)
ax[0].set_title('Grade')
ax[0].set_xticklabels(df_students.Name, rotation=90, ha='center')
```

```python
pass_counts=df_students['Pass'].value_counts()
ax[1].set_title('Passing grades')
ax[1].pie(pass_counts, labels=pass_counts)
ax[1].legend(pass_counts.keys().tolist())
fig.suptitle('Student Data')
fig.show()

#              .keys(): This method is called on a dictionary or dictionary-like object to retrieve a view object that displays a list of all the keys in the dictionary.
# .tolist(): This method is typically used to convert a NumPy array or a pandas Series to a list. However, in this context, it is likely being used on the output of .keys() to convert the view object into a list.
```
